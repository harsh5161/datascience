## Main Function

```python
import warnings
warnings.filterwarnings('ignore')
from predictive.predictive_Training import predictive_main
from segmentation.seg_Training import seg_main
from timeseries.time_Training import time_main
def main():
    '''
    PROTON MAIN FUNCTION
    '''
    print("Please select the kind of analysis you want to perform\nEnter (1) for Predictive Analysis\nEnter (2) for Segmentation Analysis\nEnter (3) for Time Series Analysis")
    analysis_input = input("Enter a number to start analysis: Enter (1) for Predictive Analysis||Enter (2) for Segmentation Analysis||Enter (3) for Time Series Analysis")
    if analysis_input == '1':
        analysis = 'Predictive'
    elif analysis_input =='2':
        analysis = 'Segmentation'
    elif analysis_input == '3':
        analysis = 'TimeSeries'
    else:
        analysis = None 

    if analysis=='Predictive':
        _ = predictive_main()
    elif analysis=='Segmentation':
       _ = seg_main()
    elif analysis=='TimeSeries':
        _ = time_main()
    else:
        _ = None
    
    if _ == None:
        print("### UNSUPPORTED OPTION SELECTED ###")
        print("### QUITTING ###")
    else:
        print("### SELECTED OPTION EXECUTED SUCCESSFULY ###")
        print("### QUITTING ###")
    return 1

if __name__ == '__main__':
    # Main Function call
    ret = main()
```
